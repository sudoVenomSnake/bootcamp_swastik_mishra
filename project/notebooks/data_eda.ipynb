# Exploratory Data Analysis

Basic EDA of raw price series.

```python

import os
import pandas as pd
import matplotlib.pyplot as plt

for i in os.listdir("../data/raw/"):
    df = pd.read_csv("../data/raw/" + i)
    df = df[2:]
    df.columns = ["Date", "Close", "High", "Low", "Open", "Volume"]
    display(df.head(), df.describe())
    df["Volume"].astype(int).plot(title = f"Volume — {i.replace(".csv", "")}")
    plt.show()
```
