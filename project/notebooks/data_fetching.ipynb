```python

import yfinance as yf

TICKERS = ["AAPL", "MSFT", "SPY"]
START = "2015-01-01"
END = None 


for t in TICKERS:
    t_fmt = t.strip().upper().replace(".", "-")
    print("Downloading", t_fmt)
    df = yf.download(t_fmt, start = START, end = END, auto_adjust = True)
    if df.empty:
        print("No data for", t_fmt)
        continue
    out = f"../data/raw/{t_fmt}.csv"
    df.to_csv(out)
    print("Saved ->", out)
```
